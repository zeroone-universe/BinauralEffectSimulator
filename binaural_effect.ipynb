# 바이노럴 효과 구현 시물레이터

본 코드는 광주과학기술원 <사운드 디자인과 프로그래밍> 기말 프로젝트 제출용으로 제작되었습니다.

필요한 라이브러리를 load합니다.

```python
import numpy as np
import sys

from PyQt5.QtWidgets import QApplication
from IPython.display import Audio

from utils import *
```

사용할 모노 음원의 path를 입력합니다.

```python
wav_path = "/Users/youngwonchoi/Desktop/code/SoundDesign/mono_wav/letsgo.wav"
```

반원형 좌표계에서 사용자와 음원의 상대적인 위치를 결정합니다.

```python
vs = QApplication(sys.argv)
pos = PolarCoordinate()
vs.exec_()

xPos, yPos=pos.xPos, pos.yPos
if xPos>=0:
    print("음원은 사용자 {0}m 앞에, {1}m 오른쪽에 존재합니다.".format(yPos,xPos))
else:
    print("음원은 사용자 {0}m 앞에, {1}m 왼쪽에 존재합니다.".format(yPos,-xPos))
```

바이노럴 효과가 반영된 음원을 제작합니다.

```python
left_vol, left_timedelay, right_vol, right_timedelay=binaural(xPos,yPos)

wav_mono, wav_left, wav_right, sr = load_audio(wav_path)
left_arraydelay=int(left_timedelay * sr)
right_arraydelay=int(right_timedelay * sr)
out_left=np.concatenate((np.zeros(left_arraydelay),(wav_left*left_vol)),axis=0)
out_left=out_left[0:len(wav_mono)+sr]
out_right=  np.concatenate((np.zeros(right_arraydelay),(wav_right*right_vol)),axis=0)  
out_right=out_right[0:len(wav_mono)+sr]

```

Input으로 사용된 모노 오디오 파일과 바이노럴 효과가 반영된 스테레오 오디오 파일을 비교해 봅니다.

```python
print("모노 오디오 파일")
Audio(wav_mono, rate = sr)
```

```python
print("스테레오 오디오 파일")
Audio([out_left, out_right], rate = sr)
```
